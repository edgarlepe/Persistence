```python
import numpy as np
import pandas as pd
import gudhi as gd
from sklearn import manifold
from pylab import *
# import sklearn_tda
a=pd.read_csv("data/DSL-StrongPasswordData.csv")
people=[a.iloc[(n*50):(n+1)*50,3:] for n in range(0,2)]

b=people[0].values.tolist()
skeleton=gd.RipsComplex(points=b, max_edge_length =.4)
simplextree_0=skeleton.create_simplex_tree(max_dimension=3)

c=people[1].values.tolist()
skeleton=gd.RipsComplex(points=c, max_edge_length =.4)
simplextree_1=skeleton.create_simplex_tree(max_dimension=3)

print(simplextree_0.num_vertices())
print(simplextree_0.num_simplices())
print(simplextree_1.num_vertices())
print(simplextree_1.num_simplices())
```

```python
BarCodes_0 = simplextree_0.persistence()
BarCodes_1 = simplextree_1.persistence()
print(len(BarCodes_0))
print(len(BarCodes_1))
```
